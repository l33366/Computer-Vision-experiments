```python
# ________________________list of comprehension method____________________
def Grayscale(img):
    height,width,ch=img.shape
    r= img[:,:,0]
    g= img[:,:,1]
    b= img[:,:,2]
    img_binary=[[0.299*b[i][j] + 0.587*g[i][j] + 0.144*r[i][j] for j in range(width)]for i in range(height)]
    return img_binary

img= cv2.imread('lenna.tif')
l1=time.time()
gscale=Grayscale(img)
l2=time.time()

LC_gray=l2-l1
plt.imshow(gscale,cmap='gray')
#___________________________________________________________________________________________
```

```python
img = image.imread('lenna.tif')

################## using for loop method ###############################
def GreyScaleCONV(img):
    height,width,channel=img.shape
    img1=np.zeros([height,width])
     
    for i in np.arange(height):
        for j in np.arange(width):
            r = img.item(i,j,0)
            g = img.item(i,j,1)
            b = img.item(i,j,2)
        
            img1[i][j] = 0.299*r + 0.587*g + 0.144*b
        
    return plt.imshow(img1,cmap='gray')

start3=time.time()
GreyScaleCONV(img)
end3=time.time()
#################################################################################
print(img.shape)
runtime3 =end3 - start3
```

```python
data = {'run_time_withforloop':runtime3, 'runtime_for_opcv2':runtime1, 'runtime_for_list_comprehension_method':LC_gray}
y_axis = list(data.keys())
x_axis = list(data.values())
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(y_axis, x_axis, color ='black',
        width = 0.2)
 
plt.xlabel("methods")
plt.ylabel("time taken to execute")
plt.title("processing time")
plt.show()
```

# Q. 6. (a)Write python from scratch for 2D Spatial Separable convolution between input image(lena.jpg) and Gausian filter = np.array ([1,4,6,4,1], [4,16,24,16,4], [6,24,36,24,6], [4,16,24,16,4], [1,4,6,4,1])/256 
(b) Compute number of multiplications required for 2D Spatial Separable convolution

```python
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib.image as image
```

```python
def shiftArr(arr, d):
    n = arr.shape[0]
    #print(n)
    arr2 = np.zeros(n)
    arr2[d:n] = arr[0:n-d]
    return arr2
```

```python
def transpose(arr):
    dim = arr.shape
    arr2 = [[arr[i][j] for i in range(dim[0])] for j in range(dim[1])]
    return arr2
```

```python
def convolve1D(X, H):
    nx = X.shape[0]
    nh = H.shape[0]
    ny = nx + nh - 1
    
    padX = np.pad(X, (0,nh-1), mode='constant')
    padH = np.pad(H, (0,nx-1), mode='constant')
    Ht0 = np.zeros((ny,ny))
    
    for i in range(ny):
        Ht0[i] = shiftArr(padH, i)
        
    Ht = transpose(Ht0)
    vX = padX.reshape(ny,1)
    Y = np.matmul(Ht, vX)
    numOfMult = Ht0.shape[0] * Ht0.shape[1]
    
    return Y.reshape(1, ny)[0],numOfMult
```

```python
def ssconvolve(x,f):
    dimx = x.shape
    nf = f.shape[0]
    dimy = (dimx[0]+nf-1,dimx[1]+nf-1)
    rowGF = f[0] * math.sqrt(1/f[0][0])
    colGF = np.array([f[i][0] * math.sqrt(1/f[0][0]) for i in range(nf)])
    
    yr = np.zeros((dimx[0],dimy[1]))
    y = np.zeros(dimy)
    numOfMult = 0
    
    for i in range(yr.shape[0]):
        yr[i],num = convolve1D(x[i], rowGF)
        numOfMult += num
    
    yc = np.array(transpose(yr))
    for i in range(yr.shape[0]):
        y[i],num = convolve1D(yc[i], colGF)
        numOfMult += num
    
    return transpose(y),numOfMult
    
```

```python
GFilter = np.array ([[1,4,6,4,1], [4,16,24,16,4], [6,24,36,24,6], [4,16,24,16,4], [1,4,6,4,1]])/256
print('Gaussian Filter:')
print(GFilter)
img = cv2.imread('lenna.tif',0)
img2 = img[0::8,0::8]
plt.imshow(img2, cmap='gray')
plt.title('Input image downscaled to 64x64')
plt.show()

convImg,numOfMult = ssconvolve(img2, GFilter) 
print(np.array(convImg).shape)
plt.imshow(convImg, cmap='gray')
plt.title('Output image after convulation with')
plt.show()
```
