# Q.1. Write python code from scratch for cany edge detection method for image (lena.tif) and demonstrate results of each step and also compare final result with OpenCV functions   compare results with OpenCV function

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import math as mt
```

## step 1: colour to grey

```python
path='lena.jpg'
img=cv2.imread(path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap='gray')
```

## step 2: gaussian blur

```python
Gausimg = cv2.GaussianBlur(img, (5,5),cv2.BORDER_DEFAULT)
plt.imshow(img,cmap='gray')
```

## step 3: get soble operators and find direction

```python
def EdgeSobleX(img):
    m,n=img.shape
    opimg1=np.zeros((m-1,n-1))
    opimg2=np.zeros((m-1,n-1))
    k1,k2=-1,1
    k11,k22=-2,2
    for i in range(1, m-1):
        for j in range(1, n-1):
            opimg1[i][j]= (k1*img[i-1][j-1] + 0*img[i-1][j] + k2*img[i-1][j+1]
                          + k11*img[i][j-1] + 0*img[i][j] + k22*img[i][j+1]
                          + k1*img[i+1][j-1] + 0*img[i+1][j] + k2*img[i+1][j+1])
    return opimg1
    
def EdgeSobleY(img):
    m,n=img.shape
    opimg1=np.zeros((m-1,n-1))
    opimg2=np.zeros((m-1,n-1))
    k1,k2=-1,1
    k11,k22=-2,2
    for i in range(1, m-1):
        for j in range(1, n-1):
            opimg2[i][j]= (k1*img[i-1][j-1] + k11*img[i-1][j] + k1*img[i-1][j+1]
                          + 0*img[i][j-1] + 0*img[i][j] + 0*img[i][j+1]
                          + k2*img[i+1][j-1] + k22*img[i+1][j] + k2*img[i+1][j+1])
    return opimg2

def Soble(img):
    Gx=EdgeSobleX(img)
    Gy=EdgeSobleY(img)
    laplacian= Gx*Gx + Gy*Gy
    m,n=laplacian.shape
    lapop=np.zeros((m,n))
    for i in range(0,m):
        for j in range(0,n):
            lapop[i][j]=mt.sqrt(laplacian[i][j])
    return lapop
Gx=EdgeSobleX(img)
Gy=EdgeSobleY(img)
G=Soble(Gausimg)
plt.imshow(G,cmap='gray')
```

```python

theta = np.arctan2(Gx, Gy)
angle = np.rad2deg(theta)
plt.imshow(angle,cmap='gray')
```

## step 4: Non max supression
#### 0 to 22.5=0 and 157.5 to 180 = 0
#### 22.5 to 67.5 =45
#### 67.5 to 112.5 =90
#### 112.5 to 157.5 =135

```python
mag = np.hypot(Gx, Gy) # Given the legs of a right triangle, return its hypotenuse.
                                      # Equivalent to sqrt(x1**2 + x2**2), element-wise.
mag = mag/mag.max()*255
mag = np.uint8(mag)
plt.imshow(mag)
```

```python
M, N = mag.shape
Non_max = np.zeros((M,N), dtype= np.uint8)
 
for i in range(1,M-1):
    for j in range(1,N-1):
       # Horizontal 0
        if (0 <= angle[i,j] < 22.5) or (157.5 <= angle[i,j] <= 180) or (-22.5 <= angle[i,j] < 0) or (-180 <= angle[i,j] < -157.5):
            b = mag[i, j+1]
            c = mag[i, j-1]
        # Diagonal 45
        elif (22.5 <= angle[i,j] < 67.5) or (-157.5 <= angle[i,j] < -112.5):
            b = mag[i+1, j+1]
            c = mag[i-1, j-1]
        # Vertical 90
        elif (67.5 <= angle[i,j] < 112.5) or (-112.5 <= angle[i,j] < -67.5):
            b = mag[i+1, j]
            c = mag[i-1, j]
        # Diagonal 135
        elif (112.5 <= angle[i,j] < 157.5) or (-67.5 <= angle[i,j] < -22.5):
            b = mag[i+1, j-1]
            c = mag[i-1, j+1]           
            
        # Non-max Suppression
        if (mag[i,j] >= b) and (mag[i,j] >= c):
            Non_max[i,j] = mag[i,j]
        else:
            Non_max[i,j] = 0

plt.figure() 
f, plots = plt.subplots(1, 2)  
plots[0].imshow(mag) 
plots[1].imshow(Non_max) 
```

## Step 5 : double thresholding 

```python
highThreshold = 21
lowThreshold = 15
 
M, N = Non_max.shape
out = np.zeros((M,N), dtype= np.uint8)
 
# If edge intensity is greater than 'High' it is a sure-edge
# below 'low' threshold, it is a sure non-edge
strong_i, strong_j = np.where(Non_max >= highThreshold)
zeros_i, zeros_j = np.where(Non_max < lowThreshold)
 
# weak edges
weak_i, weak_j = np.where((Non_max <= highThreshold) & (Non_max >= lowThreshold))
 
# Set same intensity value for all edge pixels
out[strong_i, strong_j] = 255
out[zeros_i, zeros_j ] = 0
out[weak_i, weak_j] = 75


plt.figure() 
f, plots = plt.subplots(1, 2)  
plots[0].imshow(Non_max) 
plots[1].imshow(out) 
```

## Step 6: edge tracking

```python
M, N = out.shape
for i in range(1, M-1):
    for j in range(1, N-1):
        if (out[i,j] == 75):
            if 255 in [out[i+1, j-1],out[i+1, j],out[i+1, j+1],out[i, j-1],out[i, j+1],out[i-1, j-1],out[i-1, j],out[i-1, j+1]]:
                out[i, j] = 255
            else:
                out[i, j] = 0

                
plt.imshow(out, cmap='gray') 
plt.show()
```

# Q.2. Write python code from scratch for proposed edge detection method explained by author in research paper 1  (lena.tif)  and demonstrate results of each step.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import math as mt
```

```python
path='lena.jpg'
img=cv2.imread(path)
print(img.shape)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.resize(img, (512, 512))
plt.imshow(img,cmap='gray')
```

```python
m,n=img.shape
opimg=np.zeros((m,n))
for i in range(2,m-2):
    for j in range(2,n-2):
        p=0
        avg= (img[i-2][j-2]+img[i-2][j-1]+img[i-2][j]+img[i-2][j+1]+img[i-2][j+2]
             +img[i-1][j-2]+img[i-1][j-1]+img[i-1][j]+img[i-1][j+1]+img[i-1][j+2]
             +img[i][j-2]+img[i][j-1]+img[i][j+1]+img[i][j+2]
             +img[i+1][j-2]+img[i+1][j-1]+img[i+1][j]+img[i+1][j+1]+img[i+1][j+2]
             +img[i+2][j-2]+img[i+2][j-1]+img[i+2][j]+img[i+2][j+1]+img[i+2][j+2])/24
        
        lst=[img[i-2][j-2],img[i-2][j-1],img[i-2][j],img[i-2][j+1],img[i-2][j+2]
             ,img[i-1][j-2],img[i-1][j-1],img[i-1][j],img[i-1][j+1],img[i-1][j+2]
             ,img[i][j-2],img[i][j-1],img[i][j+1],img[i][j+2]
             ,img[i+1][j-2],img[i+1][j-1],img[i+1][j],img[i+1][j+1],img[i+1][j+2]
             ,img[i+2][j-2],img[i+2][j-1],img[i+2][j],img[i+2][j+1],img[i+2][j+2]]
        
        for k in range(0,len(lst)):
            var=lst[k]-img[i][j]
            if(var<=15):
                p=p+1
        
        treshvar=abs(img[i][j]-avg)
        if((treshvar<=120) and (p>=9)):
            opimg[i][j]=255
        else:
            opimg[i][j]=0
        
     
               
plt.imshow(opimg,cmap='gray')        
        
        
```

### Denoising the img 

```python
Denoisedopimg=np.zeros((m,n))
for i in range(2,m-2):
    for j in range(2,n-2):
        if(opimg[i][j]==0):
            count=0
            lst=[opimg[i][j-1],opimg[i][j+1],opimg[i-1][j],opimg[i+1][j]]
            for k in range(0,len(lst)):
                if(lst[k]==255):
                    count=count+1
            if(count>=1):
                Denoisedopimg[i][j]=255
        else:
            count=0
            lst=[opimg[i][j-1],opimg[i][j+1],opimg[i-1][j],opimg[i+1][j]]
            for k in range(0,len(lst)):
                if(lst[k]==0):
                    count=count+1
            if(count>=4 or count>=3):
                Denoisedopimg[i][j]=0
                
            
plt.imshow(Denoisedopimg,cmap='gray')          
```
